**Preaparing the Dataset**

```python
dataset=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
dataset['Price']=boston.target
```

Independent and Dependent Features

```python
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_test
```

```python
X_train
```

Standardize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

**Model Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameter the model has been trained
regression.get_params()
```

```python
## Prediction with Test Data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot a cstter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
## Residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals
sns.displot(residuals,kind="kde")
```

```python
## Scatter Plot with rspect to prediction and residuals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# **R square and adjusted R square**

Formula
**R^2 = 1 -SSR/SST**

R^2 = coefficient of determination SSR

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

**Adjusted R2 = 1-[(1-R2)*(n-1)/(n-k-1)]**

where:

R2:The R2 of the model n: The number of observations k

```python
# display adjusted R-squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
